Análise exploratória de dados usando uma subamostra do dataset das vitimas do Titanic.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('titanic.csv')
```

```python
df['survived'] = df['survived'].replace({1:'Sim', 0:'Não'})
```

```python
df.columns
```

```python
df = df.rename(columns={'pclass': 'Classe',
                   'survived': 'Sobrevivencia', 
                   'name':'Nome',
                   'sex' : 'Sexo',
                   'age' : 'Idade',
                   'sibsp' : 'IrmaosQTD',
                   'parch' : 'Pais_Filhos',})
```

```python
df['Sexo'] = df['Sexo'].replace({'female':'Feminino', 'male':'Masculino'})
```

```python
df
```

```python
genero = df['Sexo'].value_counts().to_frame()
percent_gen = round(df['Sexo'].value_counts() / df.shape[0],1)*100
media_idade = round(df.groupby('Sexo')['Idade'].mean().to_frame(),1)
genero.columns = ['Contagem']
percent_gen.columns = ['Percentual']
resultado_gen = pd.concat([genero, percent_gen,media_idade], axis=1)
resultado_gen

```

```python
Sobrev = df['Sobrevivencia'].value_counts().to_frame()
perc_Sobrev = round((df['Sobrevivencia'].value_counts().to_frame() / df.shape[0])*100,2)
Sobrev.columns = ['Contagem']
perc_Sobrev.columns = ['Percentual']
result_sobrev = pd.concat([Sobrev, perc_Sobrev], axis=1)
result_sobrev
```

```python
cont_classe = df['Classe'].value_counts().to_frame()
cont_classe.columns = ['Contagem']
perc_classe = round((df['Classe'].value_counts().to_frame() / df.shape[0]) * 100,2)
perc_classe.columns = ['Percentual %']
resultado_classe = pd.concat([cont_classe, perc_classe], axis=1)
resultado_classe



```

```python
df.groupby(['Classe','Sexo']).agg({'Sexo' : 'count'}).rename(columns={'Sexo':'Contagem'})

```

```python
df.groupby(['Sexo','Sobrevivencia']).agg({'Sobrevivencia' : 'count'}).rename(columns={'Sobrevivencia':'Contagem'})
```

```python
plt.figure(figsize=(7,7))
sns.countplot(x='Sexo', data=df)
plt.title('Distribuição de Genero de Titanic')
plt.xlabel('Gênero')
plt.ylabel('Freuqiencia')
plt.show()
```

```python
plt.figure(figsize=(7,7))
sns.countplot(x='Classe', data=df, hue=('Sobrevivencia'))
plt.title('Distribuição de sobrevivencia por classe')
plt.xlabel('Frequencia')
plt.ylabel('Classe')
```

```python
plt.figure(figsize=(7,7))
sns.countplot(x='Sexo', data=df, hue=('Sobrevivencia'))
plt.title('Distribuição de sobrevivencia por genero')
plt.xlabel('Gênero')
plt.ylabel('Freuqiencia')
plt.show()
```

```python
sns.boxplot(x='Sobrevivencia', y='Idade', data=df)
```

```python
df.head()
```

```python
homens = df.query('Sexo == "Masculino"')
mulheres = df.query('Sexo == "Feminino"')

fig, axes = plt.subplots(1,2, figsize=(10,6))
axes[0].hist(mulheres[mulheres['Sobrevivencia']=='Sim'].Idade.dropna(),bins=20)
axes[0].hist(mulheres[mulheres['Sobrevivencia']=='Não'].Idade.dropna(),bins=20)
axes[0].set_xlabel('Idade')
axes[0].legend(['Sobreviventes','Mortos'])
axes[0].set_title('Mulheres')
axes[1].hist(homens[homens['Sobrevivencia']=='Sim'].Idade.dropna(),bins=20)
axes[1].hist(homens[homens['Sobrevivencia']=='Não'].Idade.dropna(),bins=20)
axes[1].set_xlabel('Idade')
axes[1].legend(['Sobreviventes','Mortos'])
axes[1].set_title('Homens')

```

```python
corr_matriz = df.select_dtypes(include='number').corr()
```

```python
corr_matriz
```

```python
plt.figure(figsize=(12,7))
sns.heatmap(corr_matriz,cmap='RdBu_r', vmin=-1, vmax=1, annot=True)
plt.title('Matriz de Correlação', fontsize=10)
plt.show()
```

```python
porto_s = df.query('embarked == "S"')
porto_c = df.query('embarked == "C"')
porto_q = df.query('embarked == "Q"')
fig, axes = plt.subplots(1, 2, figsize=(10, 6))

# Porcentagem de sobreviventes
axes[0].bar(['S', 'C', 'Q'], [
    (porto_s[porto_s['Sobrevivencia'] == 'Sim'].Sobrevivencia.count() / porto_s['Sobrevivencia'].count()) * 100,
    (porto_c[porto_c['Sobrevivencia'] == 'Sim'].Sobrevivencia.count() / porto_c['Sobrevivencia'].count()) * 100,
    (porto_q[porto_q['Sobrevivencia'] == 'Sim'].Sobrevivencia.count() / porto_q['Sobrevivencia'].count()) * 100
])
axes[0].set_title('Porcentagem de Sobreviventes')

# Porcentagem de mortos
axes[1].bar(['S', 'C', 'Q'], [
    (porto_s[porto_s['Sobrevivencia'] == 'Não'].Sobrevivencia.count() / porto_s['Sobrevivencia'].count()) * 100,
    (porto_c[porto_c['Sobrevivencia'] == 'Não'].Sobrevivencia.count() / porto_c['Sobrevivencia'].count()) * 100,
    (porto_q[porto_q['Sobrevivencia'] == 'Não'].Sobrevivencia.count() / porto_q['Sobrevivencia'].count()) * 100
])
axes[1].set_title('Porcentagem de Mortos')

# Ajustando layout
plt.tight_layout()
plt.show()

```
