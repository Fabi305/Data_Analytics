## Visualização de dados

```python
# Cálculo dos KPIs
media_idade = df['idade'].mean()
salario_medio = df['salario_anual'].mean()
gasto_eletronicos = df['gasto_com_eletronicos'].sum()
total_compras_web = df['numero_de_compras_na_web'].sum()
media_filhos = df['filhos_em_casa'].mean()
media_visitas_website = df['numero_visitas_website_mes'].mean()
total_compras_catalogo = df['numero_de_compras_via_catalogo'].sum()
total_compras_desconto = df['numero_de_compras_com_desconto'].sum()

# Lista de KPIs
kpis = [
    {"title": "Média de Idade", "value": round(media_idade, 2)},
    {"title": "Salário Anual Médio", "value": f"${round(salario_medio, 2):,}"},
    {"title": "Total Gasto com Eletrônicos", "value": f"${round(gasto_eletronicos, 2):,}"},
    {"title": "Total de Compras na Web", "value": total_compras_web},
    {"title": "Média de Filhos em Casa", "value": round(media_filhos, 2)},
    {"title": "Média de Visitas ao Website", "value": round(media_visitas_website, 2)},
    {"title": "Total de Compras via Catálogo", "value": total_compras_catalogo},
    {"title": "Total de Compras com Descontos", "value": total_compras_desconto}
]

# Criação dos gráficos dos KPIs
fig, axs = plt.subplots(3, 3, figsize=(15, 9))  # 3 linhas e 3 colunas

# Desativar eixos e criar cartões para cada KPI
for i, ax in enumerate(axs.flat):
    if i < len(kpis):
        ax.axis('off')
        ax.text(0.5, 0.7, kpis[i]["title"], fontsize=15, ha='center', va='center', color='gray', weight='bold')
        ax.text(0.5, 0.3, str(kpis[i]["value"]), fontsize=30, ha='center', va='center', color='black', weight='bold')
        ax.spines['top'].set_visible(True)
        ax.spines['right'].set_visible(True)
        ax.spines['bottom'].set_visible(True)
        ax.spines['left'].set_visible(True)
        
        # Adiciona bordas
        for spine in ax.spines.values():
            spine.set_edgecolor('black')
            spine.set_linewidth(2)
    else:
        ax.axis('off')  # Esconde os subplots vazios

plt.tight_layout()
plt.show()
```

```python
categoria_pais = df.groupby('pais')[[                    
                    'gasto_com_eletronicos',
                    'gasto_com_brinquedos',
                    'gasto_com_moveis',
                    'gasto_com_utilidades',	
                    'gasto_com_alimentos',
                    'gasto_com_vestuario']].sum().reset_index()
categoria_pais.columns = ['País','Eletronicos', 'Brinquedos', 'Moveis', 'Utilidades', 'Alimentos', 'Vestuario']
#formato longo
gastos_long = pd.melt(categoria_pais, 
                      id_vars='País', 
                      var_name='Categoria', 
                      value_name='Total Gasto') 

plt.figure(figsize = (14,8))
sns.barplot(data=gastos_long,
            x='País',
            y= 'Total Gasto',
            hue='Categoria',
            alpha = 1)
plt.title('Total de Gastos em Diferentes Categorias', 
          fontsize=14,
          weight = 'bold')
plt.show()
```

```python

df['ano'] = df['data_cadastro'].dt.year
total_gasto_ano = df.groupby(['ano','pais'])['total_gasto'].sum().reset_index()
total_gasto_ano.columns = ['Ano','país','Total Gasto']

plt.figure(figsize=(12,7))
sns.lineplot(data=total_gasto_ano,
             x='Ano',
             y='Total Gasto',
             hue = 'país',
             marker='o',
             palette = 'plasma')
plt.title('Total Gasto Por ano e País', 
          fontsize=14,
          weight='bold')
plt.show()
```
